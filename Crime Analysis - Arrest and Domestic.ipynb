## Analyzing relationship between Arrests and Domestic Crimes 

```python
# This is the sum of the number of arrests and number of domestic crimes
crimes[["Arrest","Domestic"]].sum()
```

```python
# Now in percentage of the total lines
crimes[["Arrest","Domestic"]].sum()/crimes.shape[0]
```

We can verify a small number of arrests: only 1/5 of the whole.

Let's see how they break down when we mix both:

```python
# Breakdown of crimes by domestic and suspected arrested, as a percentage of the total:
arrests_domestic = (crimes[["ID","Arrest","Domestic"]].groupby(["Arrest","Domestic"]).count()/
                    crimes.shape[0])
arrests_domestic
```

```python
arrests_domestic_2 = (crimes[["ID","Domestic","Arrest"]].pivot_table(columns=["Arrest","Domestic"], aggfunc='count')
                     /crimes.shape[0])
arrests_domestic_2
```

```python
# Absolute number of domestic crimes that have resulted in arrest
crimes.pivot_table(values="Arrest",columns=["Domestic"], aggfunc='sum')
```

```python
# Relative number of non-domestic and domestic crimes that have resulted in arrest
arrests-rate_by_domestic = (crimes.pivot_table(values="Arrest",columns=["Domestic"], aggfunc='sum')/
                             crimes.pivot_table(values="Arrest",columns=["Domestic"], aggfunc='count'))

arrests-rate_by_domestic
```

#### We can conclude that there's a small difference between the rate of arrestations per Domestic variable. Can we verify the statistical significance of that ?

## Analyzing Beats

- **Beat** - Indicates the beat where the incident occurred. A beat is the smallest police geographic area – each beat has a dedicated police beat car. Three to five beats make up a police sector, and three sectors make up a police district. The Chicago Police Department has 22 police districts. See the beats at https://data.cityofchicago.org/d/aerh-rz74.
